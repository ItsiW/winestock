# Example workflow

````python
site = "Berkeley Bowl"

wine_df = get_latest_scrape(site)
wine_df = flag_errors(wine_df)

wine_df["discount"] = wine_df.price.divide(wine_df.vivino_price.fillna(wine_df.price))
````

````python
view_links(
    wine_df[
        (wine_df.vivino_score > 3.8)
        & (wine_df.num_reviews > 50)
    ]
    .sort_values("discount")
    .head(20)
)
````
